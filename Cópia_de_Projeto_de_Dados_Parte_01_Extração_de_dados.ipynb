### 1. Carregar os dados do arquivo CSV

Este código importa a biblioteca `pandas` e carrega o arquivo `PremierLeague-2324.csv` em um DataFrame. Ele também exibe as primeiras linhas e um resumo estatístico.

```python
import pandas as pd

df = pd.read_csv('PremierLeague-2324.csv')
display(df.head())
display(df.describe())
```

### 2. Traduzir os nomes das colunas

Este bloco de código define um dicionário de tradução e o aplica ao DataFrame para renomear as colunas para o português.

```python
column_translation = {
    'Date': 'Data',
    'HomeTeam': 'TimeCasa',
    'AwayTeam': 'TimeFora',
    'FTHG': 'GolsTimeCasaFT',
    'FTAG': 'GolsTimeForaFT',
    'FTR': 'ResultadoFinal',
    'HTHG': 'GolsTimeCasaHT',
    'HTAG': 'GolsTimeForaHT',
    'HTR': 'ResultadoIntervalo',
    'Referee': 'Juiz',
    'HS': 'ChutesTimeCasa',
    'AS': 'ChutesTimeFora',
    'HST': 'ChutesNoAlvoTimeCasa',
    'AST': 'ChutesNoAlvoTimeFora',
    'HF': 'FaltasTimeCasa',
    'AF': 'FaltasTimeFora',
    'HC': 'EscanteiosTimeCasa',
    'AC': 'EscanteiosTimeFora',
    'HY': 'CartoesAmarelosTimeCasa',
    'AY': 'CartoesAmarelosTimeFora',
    'HR': 'CartoesVermelhosTimeCasa',
    'AR': 'CartoesVermelhosTimeFora'
}

df = df.rename(columns=column_translation)
display(df.head())
```

```python
pd.set_option('display.max_columns', None)
df.head(380)
```

###Top 10 Times com Mais Gols na Temporada
A tabela abaixo apresenta os 10 times que mais marcaram gols na temporada, exibidos em um formato horizontal para facilitar a visualização e comparação direta dos totais de gols entre as equipes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate total goals scored by home teams
home_goals = df.groupby('TimeCasa')['GolsTimeCasaFT'].sum().reset_index()
home_goals.columns = ['Time', 'GolsMarcados']

# Calculate total goals scored by away teams
away_goals = df.groupby('TimeFora')['GolsTimeForaFT'].sum().reset_index()
away_goals.columns = ['Time', 'GolsMarcados']

# Combine home and away goals for total goals scored by each team
total_goals = pd.concat([home_goals, away_goals]).groupby('Time')['GolsMarcados'].sum().reset_index()
total_goals.columns = ['Time', 'GolsTotais']

# Sort the teams by total goals in descending order
total_goals = total_goals.sort_values(by='GolsTotais', ascending=False)

# Get the top 10 teams by total goals
top_10_teams = total_goals.head(10)

# Create the bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x='Time', y='GolsTotais', data=top_10_teams, hue='Time', palette='viridis', legend=False)
plt.title('Top 10 Times com Mais Gols na Temporada', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Total de Gols', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
